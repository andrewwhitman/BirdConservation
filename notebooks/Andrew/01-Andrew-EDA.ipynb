```python
# Diet columns
diets = [c for c in df.columns if re.search('Diet.+', c)]
diets
```

```python
# check that proportions sum to 100
diet_prop = diets[:10]
(df[diet_prop].sum(axis=1) == 100.0).all()
```

```python
# ForStrat columns
strats = [c for c in df.columns if re.search('ForStrat.+', c)]
strats
```
